```python
# convert the dates from a string to a python DateTime type
from datetime import datetime

def pd_to_datetime(date_value):
    return pd.to_datetime(date_value) if not pd.isna(date_value) else  datetime(1970,1,1,0,0)

sentenceDS['Con victDate'] = sentenceDS['Con victDate'].apply(pd_to_datetime)
sentenceDS['StartDate'] = sentenceDS['StartDate'].apply(pd_to_datetime)
sentenceDS['EndDate'] = sentenceDS['EndDate'].apply(pd_to_datetime)
```

```python

def get_days (SentCode):
    if type(SentCode) == str:
        code_parts = SentCode.split(' ')
    else:
        return 0
    
    years = 0
    months = 0
    days = 0
    for idx,val in enumerate(code_parts):
        if val == 'Y':
            years = int(code_parts[idx-1])
        elif val == 'M':
            months = int(code_parts[idx-1])
        elif val == 'D':
            days = int(code_parts[idx-1])
    return (years*365) + (months *30) + days


sentenceDS['Days'] = sentenceDS['SentenceTerm'].apply(get_days)
```
