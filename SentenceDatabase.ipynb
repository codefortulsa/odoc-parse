```python
import numpy as np
import pandas as pd


sentenceDS = pd.read_csv('data/OffenderSentence.csv', encoding='iso-8859-1')

sentenceDS.head()
```

```python
# convert the dates from a string to a python DateTime type
from datetime import datetime

def pd_to_datetime(date_value):
    return pd.to_datetime(date_value) if not pd.isna(date_value) else  datetime(1970,1,1,0,0)

sentenceDS['Con victDate'] = sentenceDS['Con victDate'].apply(pd_to_datetime)
sentenceDS['StartDate'] = sentenceDS['StartDate'].apply(pd_to_datetime)
sentenceDS['EndDate'] = sentenceDS['EndDate'].apply(pd_to_datetime)
```

```python
import re 

find_parts = re.compile(r'^^((?P<years>\d+)\sY)?\s?((?P<months>\d+)\sM)?\s?((?P<days>\d+)\sD)?')

def get_parts (SentCode):
    if type(SentCode) == str:
        parts = find_parts.match(SentCode).groupdict()
        return (parts['years'], parts['months'], parts['days'])
    else:
        return (0, 0, 0)

# split the Sentence Code into Y M D and assign to columns
sentenceDS['Years'], sentenceDS['Months'], sentenceDS['Days'] = np.vectorize(get_parts)(sentenceDS['SentenceTerm'])
```

```python
def create_index(court_str, crf_num):
#   Figure out the county
    court_words = court_str.split(' ')
    county = court_words[0].lower()
    case_type = 'CF'
    
    crf_parts = crf_num.split('-')
    
#   Figure out the case number
    if len(crf_parts) == 4:
        county_abbrv, case_type, year, case_number = crf_parts
    elif len(crf_parts) == 3:
        case_type, year, case_number = crf_parts
    elif len(crf_parts) == 2:
        year, case_number = crf_parts
    else:
        year = '0000'
        case_number = crf_parts[0]
        
# Fix out the year
    year = re.sub('[\D\s]','',year)
    
    if len(year) == 2:
        year = '20'+year if int(year)<20 else '19'+year
    elif len(year) == 1:
        year = '200'+year

        
    oscn_index  = '-'.join([county,'CF',year,case_number])
    return oscn_index
    
sentenceDS['OSCN_Case_Index'] = np.vectorize(create_index)(sentenceDS['Court'], sentenceDS['CRFNum'])
```
