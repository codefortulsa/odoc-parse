```python
import csv
import sys
import pandas as pd

csv.field_size_limit(sys.maxsize)
sentence_database = []
first_row = 0

with open('OffenderSentence.csv', encoding = "ISO-8859-1") as csvfile:
    offender_reader = csv.reader(csvfile, delimiter=',')
    for row in offender_reader:
        if first_row == 0:
            print(row)
            first_row = row
        else:
            sentence_database.append(row)
```

```python
sentenceDS = pd.DataFrame(data=sentence_database, columns=first_row)
```

```python
pd.pivot_table(sentenceDS, values=["DOCNum"], index=["SentenceTermCode"], aggfunc="count")
```

```python
sentenceDS[sentenceDS.SentenceTermCode == ""].pivot_table(values=["DOCNum"], index=["SentenceTerm"], aggfunc="count")
```
